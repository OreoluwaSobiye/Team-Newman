```python
#import selenium webdriver 
from selenium import webdriver
from selenium.webdriver.common.by import By #import By
from selenium.webdriver.support.ui import WebDriverWait #import time delay
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.action_chains import ActionChains

# launch chrome browser for automation and assign to a variable
driver = webdriver.Chrome()

# Get selenium to open the konga url
driver.get("https://www.konga.com/")

# Find the login button
my_element = driver.find_element(By.XPATH, '//*[@id="nav-bar-fix"]/div[1]/div/div[4]/a')

# Click the login button
my_element.click()

#Fill in login details
username = driver.find_element(By.ID, "username")
password = driver.find_element(By.ID, "password")
username.send_keys("mercysobiye@gmail.com")
password.send_keys("n6@D.ZEnvQxa73E")

# Login
Login = driver.find_element(By.XPATH, '//*[@id="app-content-wrapper"]/div[4]/section/section/div/aside/div[2]/div/form/div[3]/button')
Login.click()


#
cookies = driver.find_element(By.XPATH, "/html/body")
cookies.click()

#From the categories click the Computers and Accessories link
Category = WebDriverWait(driver, 10).until(
    EC.element_to_be_clickable((By.XPATH, '//*[@id="nav-bar-fix"]/div[2]/div/a[1]'))
)
Category.click()

comp_acc =  WebDriverWait(driver, 10).until(
    EC.element_to_be_clickable((By.XPATH, '//*[@id="subFixId"]/div/div[1]/ul/li[1]'))
)
comp_acc.click()

```

```python
# proceed to checkout
driver.implicitly_wait(3)
checkout = driver.find_element(By.XPATH, '//*[@id="mainContent"]/div/div[2]/div[3]/div[3]/div/div[2]/a')
driver.execute_script("arguments[0].scrollIntoView(true);", checkout)
driver.execute_script("arguments[0].click();", checkout)
```

```python
#Add and select an address
address = driver.find_element(By.XPATH, '//*[@id="mainContent"]/div/div[3]/div[1]/section[1]/div/div/div[2]/div[1]/div[2]/div[1]/div/button')
address.click()

driver.implicitly_wait(10)
add = driver.find_element(By.XPATH, '//*[@id="app-content-wrapper"]/div[2]/section/section/div/aside/div[2]/div/div/div[2]/div[1]/form/button')
add.click()

driver.implicitly_wait(10)
select = driver.find_element(By.XPATH, '//*[@id="app-content-wrapper"]/div[2]/section/section/div/aside/div[3]/div/div/div/a')
select.click()
```

```python
# Click the pay now box
pay_now = WebDriverWait(driver, 10).until(
    EC.element_to_be_clickable((By.XPATH, "//*[@id='mainContent']//span/input"))
)
pay_now.click()
```

```python
# Continue to payment
payment = driver.find_element(By.XPATH, '//*[@id="mainContent"]/div/div[3]/div[1]/section[2]/div/div[2]/div[3]/div[2]/div/button')
payment.click()

```

```python
# Switch to the payment iframe
iframe = WebDriverWait(driver, 15).until(
    EC.presence_of_element_located((By.TAG_NAME, "iframe"))
)
driver.switch_to.frame(iframe)

# Select card payment method
card = WebDriverWait(driver, 10).until(
    EC.element_to_be_clickable((By.XPATH, '//*[@id="channel-template"]//button'))
)
card.click()


# Filling card details
card_no = driver.find_element(By.ID, 'card-number')
card_date = driver.find_element(By.ID, 'expiry')
cvv = driver.find_element(By.ID, 'cvv')
card_no.send_keys('2345678575933765')
card_date.send_keys('0329')
cvv.send_keys('341')

# Locate the error message displayed for invalid card number
driver.implicitly_wait(10)
error_message = driver.find_element(By.XPATH, '//*[@id="card-number_unhappy"]')

# Print the error message text to the console
print("Payment Error Message:", error_message.text)
```

```python
# Exit the payment iframe
driver.switch_to.default_content()

# Close browser
driver.quit()
```
